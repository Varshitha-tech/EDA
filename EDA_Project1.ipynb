```python
# Try specifying the delimiter explicitly
df_country = pd.read_excel("/content/Copy of Country-Code.xlsx")
df_country.head()
# Alternatively, inspect the CSV file to identify and resolve any inconsistencies
# Once resolved, you can try reading the file again
```

```python
final_df = pd.merge(df,df_country,on='Country Code' , how= 'left')
final_df.head()
```

```python
country_values = final_df.Country.value_counts().values
```

```python
country_names = final_df.Country.value_counts().index
```

**Pie chart for top 5 countries uses zomata**

```python
plt.pie(country_values[:5],labels=country_names[:5],autopct = '%1.2f%%')
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```


when rating is b/w 4.0 to4.9 is excellent
